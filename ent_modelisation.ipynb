```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Génération de données aléatoires
np.random.seed(0)
X = np.random.rand(150, 2)

# Entraînement du modèle KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Prédiction des clusters
labels = kmeans.predict(X)

# Affichage des données et des clusters prédits
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits

# Chargement des données MNIST
digits = load_digits()
X = digits.data

# Entraînement du modèle KMeans
kmeans = KMeans(n_clusters=10)
kmeans.fit(X)

# Prédiction des clusters
labels = kmeans.predict(X)

# Affichage des données et des clusters prédits
fig, ax = plt.subplots(2, 5, figsize=(8, 3))
for axi, center in zip(ax.flat, kmeans.cluster_centers_):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center.reshape(8, 8), interpolation='nearest', cmap=plt.cm.binary)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Génération de données aléatoires
np.random.seed(0)
X = np.random.rand(150, 2)

# Normalisation des données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Entraînement du modèle KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_scaled)

# Prédiction des clusters
labels = kmeans.predict(X_scaled)

# Affichage des données et des clusters prédits
plt.scatter(X_scaled[:, 0], X)
```
